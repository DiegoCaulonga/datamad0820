```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# CHALLENGE 1

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
plt.plot(x,y)
plt.ylabel("diego")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,ls = "--",color="tomato")
plt.subplot(1,2,2)
plt.plot(x,z,color = "000000", ls="dotted",lw=3)
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(x,z)
```

```python
fig,axes = plt.subplots(1,2,figsize=(8,2))
axes[0].plot(x,y,lw=5)
axes[1].plot(x,z,ls=":")

axes[1].set_ylabel("irene")
axes[0].set_xlabel("diego")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2,x, np.exp(x))
axes[0].set_title("Escala Normal")

axes[1].plot(x, x**2,x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala logartmica")
```
